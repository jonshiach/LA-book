````python
import sympy as sp
import subprocess

class System:

    def __init__(self, A, b, label):
        self.A = A
        self.b = b
        self.label = label


def inverseSolve(system, string):

    A, b, label = system.A, system.b, system.label
    x = A.inv() * b
    string += f"{label}\n\n"
    string += "$$ \\begin{align*}\n"
    string += f"    \\det{sp.latex(A, mat_delim='(', mat_str='matrix')} &= {A.det()}, \\\ \n"
    string += f"    \\operatorname{{adj}}{sp.latex(A, mat_delim='(', mat_str='matrix')} &= {sp.latex(A.inv() * A.det(), mat_delim='(', mat_str='matrix')}, \\\ \n"
    string += f"    \\therefore A^{{-1}} &= {sp.latex(1/A.det())}{sp.latex(A.inv() * A.det(), mat_delim='(', mat_str='matrix')} = {sp.latex(A.inv(), mat_delim='(', mat_str='matrix')} \n"
    string += "\\end{align*} $$ \n\ntherefore \n\n$$\\begin{align*}\n"
    string += f"    \\vec{{x}} &= {sp.latex(A.inv(), mat_delim='(', mat_str='matrix')}\n"
    string += f"    {sp.latex(b, mat_delim='(', mat_str='matrix')} = \n"
    string += f"    {sp.latex(A.inv() * b, mat_delim='(', mat_str='matrix')}\n"
    string += "\\end{align*} $$\n\nso the solution is "
    for i in range(len(x)):
        string += f"$x_{i+1} = {sp.latex(x[i])}$"
        if i < len(x) - 2:
            string += ", "
        elif i == len(x) - 2:
            string += " and "
    string+= ".\n\n"

    return string

systems = []
systems.append(System(sp.Matrix([[-4, 2], [3, 4]]), sp.Matrix([-22, 11]), "(a)"))
systems.append(System(sp.Matrix([[-4, 2], [-1, -3]]), sp.Matrix([6, -2]), "(b)"))
systems.append(System(sp.Matrix([[-4, -4, -2], [3, 0, 4], [1, 0, 0]]), sp.Matrix([16, -8, 0]), "(c)"))
systems.append(System(sp.Matrix([[4, 0, -4], [4, -1, 1], [3, 1, 2]]), sp.Matrix([8, -4, -12]), "(d)"))

string = "```{solution} systems-ex-inverse-sol\n\n"
for i in range(len(systems)):
    string = inverseSolve(systems[i], string)

string += "```"

print(string)
subprocess.run("pbcopy", text=True, input=string)
````

````python
import sympy as sp
import subprocess

class System:

    def __init__(self, A, b, label):
        self.A = A
        self.b = b
        self.label = label


def cramer(system, string):

    A, b, label = system.A, system.b, system.label
    x = A.inv() * b
    
    string += f"{label} \n\n$$ \\begin{{align*}} \n"
    for i in range(len(x)):
        Ai = A.copy()
        Ai[:,i] = b
        string += f"    x_{{{i+1}}} &= \\frac{{\det{sp.latex(Ai, mat_delim='(', mat_str='matrix')}}}{{\det{sp.latex(A, mat_delim='(', mat_str='matrix')}}} \n"
        string += f"    = \\frac{{{sp.latex(Ai.det())}}}{{{sp.latex(A.det())}}} = {sp.latex(x[i])}"

        if i < len(x) - 1:
            string += ", \\\ \n"
        else:
            string += ". \n"

    string += "\\end{align*} $$ \n\n"
    return string

systems = []
systems.append(System(sp.Matrix([[1, 4], [-4, 1]]), sp.Matrix([-20, -5]), "(a)"))
systems.append(System(sp.Matrix([[1, 1], [0, 4]]), sp.Matrix([4, 12]), "(b)"))
systems.append(System(sp.Matrix([[3, -4, -4], [-2, -1, -1], [4, -1, 3]]), sp.Matrix([21, 8, -14]), "(c)"))
systems.append(System(sp.Matrix([[4, 4, 1], [-2, 1, 1], [-5, 4, 2]]), sp.Matrix([5, -1, -14]), "(d)"))

string = "```{solution} systems-ex-cramer\n\n"
for i in range(len(systems)):
    string = cramer(systems[i], string)

string += "```"

print(string)


subprocess.run("pbcopy", text=True, input=string)
````

````python
import sympy as sp
import subprocess

class System:

    def __init__(self, A, b, label):
        self.A = A
        self.b = b
        self.label = label


def print_augmented_matrix(A, string):
    m, n = A.shape
    cols = "c" * (n - 1)
    cols += "|c"
    string += f"    & \\left( \\begin{{array}}{{{cols}}} \n"
    for i in range(m):
        string += "        "
        for j in range(n - 1):
            string += f" {sp.latex(A[i,j])} &"
        string += f" {sp.latex(A[i,-1])} \\\ \n"
    string += "    \\end{array} \\right) \n"

    return string


def gelim(A, string):
    m, n = A.shape
    
    string += "\n$$ \\begin{align*} \n"
    string = print_augmented_matrix(A, string)
    andflag = True
    
    pivotrow = 0
    for j in range(n):

        if pivotrow == m - 1:
            break 

        # Swap pivot row if pivot is zero
        if A[pivotrow,j] == 0:  
            for i in range(pivotrow, m):
                if A[i,j] != 0:
                    break
            
            string += "\\begin{array}{l} "
            for k in range(m):
                if k == pivotrow:
                    string += f"R_{{pivotrow}} \\leftrightarrow R_{{i}} \\\ "
                else:
                    string += "\\phantom{x} "
            
            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

            continue

        # Row reduce rows below pivot row 
        if pivotrow < m - 1:
            string += "    \\begin{array}{l} "
            for i in range(m):
                scalar = A[i,j] / A[pivotrow,j]
                if i <= pivotrow:
                    string += "\\phantom{x} "
                elif scalar == 1:
                    string += f"R_{{{i+1}}} - R_{{{pivotrow+1}}} "
                elif scalar == -1:
                    string += f"R_{{{i+1}}} + R_{{{pivotrow+1}}} "
                elif scalar > 0:
                    string += f"R_{{{i+1}}} - {sp.latex(abs(scalar))} R_{{{pivotrow+1}}} "
                elif scalar < 0:
                    string += f"R_{{{i+1}}} + {sp.latex(abs(scalar))} R_{{{pivotrow+1}}}"
                else:
                    string += "\\phantom{x} "
                
                if i < m - 1:
                    string += "\\\ "

                if i <= pivotrow:
                    continue

                A[i,:] -= scalar * A[pivotrow,:]
            
            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

        # Print updated matrix
        string += "    \\longrightarrow \n"
        string = print_augmented_matrix(A, string)
        pivotrow += 1

    string += "\\end{align*} $$ \n"
                
    return A, string



def back_substitution(A, string):
    m, n = A.shape
    x = A[:,-1]
    string += "\nSolving by back substitution gives\n\n"
    string += "$$ \\begin{align*} \n"

    for i in range(m - 1, -1, -1):
        string += f"    x_{{{i+1}}} &= "
        if A[i,i] == 1:
            string += f"{sp.latex(A[i,-1])}"
        elif A[i,i] == -1:
            string += f" - \\left( {sp.latex(A[i,-1])}"
        elif A[i,i] < 0:
            string += f" - \\frac{{1}}{{{sp.latex(-A[i,i])}}} \\left( {sp.latex(A[i,-1])}"
        else:
            string += f" \\frac{{1}}{{{sp.latex(A[i,i])}}} \\left( {sp.latex(A[i,-1])}"
            
        for j in range(i + 1, n - 1):
            if A[i,j] >= 0:
                string += f" - {sp.latex(A[i,j])} \\left( {sp.latex(x[j])} \\right)"
            else:
                string += f" - \\left( {sp.latex(A[i,j])} \\right) \\left( {sp.latex(x[j])} \\right)"
            x[i] -= A[i,j] * x[j]
        
        x[i] /= A[i,i]
        
        if A[i,i] == 1:
            string += " = "
        else:
            string += " \\right) = "
        string += f"{sp.latex(x[i])}"
            
        if i > 0:
            string += ", \\\ \n"
        else:
            string += ". \n"
    
    string += "\\end{align*} $$ \n"

    return string
    

# Define systems
systems = []
systems.append(System(sp.Matrix([[-1, 3], [-2, 1]]), sp.Matrix([[-2], [1]]), "(a)"))
systems.append(System(sp.Matrix([[3, 1, 2], [4, 0, -4], [4, -2, 1]]), sp.Matrix([[11], [-4], [13]]), "(b)"))
systems.append(System(sp.Matrix([[-1, -5, -2], [2, -2, -3], [3, -1, 4]]), sp.Matrix([[-17], [-14], [-13]]), "(c)"))
systems.append(System(sp.Matrix([[-1, -5, -2], [2, -2, -3],  [3, -1, -4]]), sp.Matrix([[-26], [-19], [-20]]), "(d)"))
systems.append(System(sp.Matrix([[3, -5, -4, -1], [0, -4, 3, -4], [2, 3, 3, -3], [-2, 2, -5, -4]]), sp.Matrix([[28], [41], [11], [-21]]), "(e)"))
systems.append(System(sp.Matrix([[2, -3, -3, 4], [4, -5, 1, -5], [3, 3, -1, -5], [1, 0, 1, 3]]), sp.Matrix([[-1], [42], [20], [-4]]), "(f)"))

# Generate solution string
string = "```{solution} systems-ex-gelim\n"  

for sys in systems:
    string += f"\n{sys.label} \n"
    A = sys.A
    Ab = A.row_join(sys.b)
    Ab, string = gelim(Ab, string)
    string = back_substitution(Ab, string)

string += "\n```"
print(string)

subprocess.run("pbcopy", text=True, input=string)
````

````python
def gelimpp(A, string):
    m, n = A.shape
    
    string += "\n$$ \\begin{align*} \n"
    string = print_augmented_matrix(A, string)
    andflag = True
    
    pivotrow = 0
    for j in range(n):

        if pivotrow == m - 1:
            break 

        # Partial pivoting
        maxpivot, maxpivotrow = A[pivotrow,j], pivotrow
        for i in range(pivotrow, m):
            if abs(A[i,j]) > abs(maxpivot):
                maxpivot = A[i,j]
                maxpivotrow = i
        
        if maxpivotrow != pivotrow:
            string += "    \\begin{array}{l} "
            for k in range(m):
                if k == maxpivotrow:
                    string += f"R_{{{pivotrow+1}}} \\leftrightarrow R_{{{maxpivotrow+1}}} \\\ "
                else:
                    string += "\\phantom{x} "
                if k < m - 1:
                    string += "\\\ "
            
            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

            A[pivotrow, :], A[maxpivotrow, :] = A[maxpivotrow, :], A[pivotrow, :]
        
            # Print updated matrix
            string += "    \\longrightarrow \n"
            string = print_augmented_matrix(A, string)

        if A[pivotrow, j] == 0:
            continue

        # Row reduce rows below pivot row 
        if pivotrow < m - 1:
            string += "    \\begin{array}{l} "
            for i in range(m):
                scalar = A[i,j] / A[pivotrow,j]
                if i <= pivotrow:
                    string += "\\phantom{x} "
                elif scalar == 1:
                    string += f"R_{{{i+1}}} - R_{{{pivotrow+1}}} "
                elif scalar == -1:
                    string += f"R_{{{i+1}}} + R_{{{pivotrow+1}}} "
                elif scalar > 0:
                    string += f"R_{{{i+1}}} - {sp.latex(abs(scalar))} R_{{{pivotrow+1}}} "
                elif scalar < 0:
                    string += f"R_{{{i+1}}} + {sp.latex(abs(scalar))} R_{{{pivotrow+1}}}"
                else:
                    string += "\\phantom{x} "
                
                if i < m - 1:
                    string += "\\\ "

                if i <= pivotrow:
                    continue

                A[i,:] -= scalar * A[pivotrow,:]
            
            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

        # Print updated matrix
        string += "    \\longrightarrow \n"
        string = print_augmented_matrix(A, string)
        pivotrow += 1

    string += "\\end{align*} $$ \n"
                
    return A, string

# Generate solution string
string = "```{solution} systems-ex-gelimpp\n"  

for sys in systems:
    string += f"\n{sys.label} \n"
    A = sys.A
    Ab = A.row_join(sys.b)
    Ab, string = gelimpp(Ab, string)
    string = back_substitution(Ab, string)

string += "\n```"
print(string)

subprocess.run("pbcopy", text=True, input=string)
````

````python
def gjelim(A, string):
    m, n = A.shape
    
    string += "\n$$ \\begin{align*} \n"
    string = print_augmented_matrix(A, string)
    andflag = True
    
    pivotrow = 0
    for j in range(n):

        if pivotrow == m:
            break 

        # Swap pivot row if pivot is zero
        if A[pivotrow,j] == 0:  
            for i in range(pivotrow, m):
                if A[i,j] != 0:
                    break
            
            string += "\\begin{array}{l} "
            for k in range(m):
                if k == pivotrow:
                    string += f"R_{{pivotrow}} \\leftrightarrow R_{{i}} \\\ "
                else:
                    string += "\\phantom{x} "
            
            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

            continue

        # Divide pivot row by pivot
        if A[pivotrow,j] != 1:
            string += "    \\begin{array}{l} "
            for i in range(m):
                if i != pivotrow:
                    string += "\\phantom{x} "
                else:
                    string += f"{sp.latex(1 / A[pivotrow,j])} R_{{{pivotrow+1}}}"
            
                if i < m - 1:
                    string += "\\\ "

            if andflag and n < 5:
                string += "\\end{array} & \n"
                andflag = False
            else:
                string += "\\end{array} \\\ \\\ \n"
                andflag = True

            A[pivotrow,:] /= A[pivotrow,j]
        
            # Print updated matrix
            string += "    \\longrightarrow \n"
            string = print_augmented_matrix(A, string)

        # Row reduce rows
        string += "    \\begin{array}{l} "
        for i in range(m):
            scalar = A[i,j] / A[pivotrow,j]
            if i == pivotrow or A[i,j] == 0:
                string += "\\phantom{x} "
            elif scalar == 1:
                string += f"R_{{{i+1}}} - R_{{{pivotrow+1}}} "
            elif scalar == -1:
                string += f"R_{{{i+1}}} + R_{{{pivotrow+1}}} "
            elif scalar > 0:
                string += f"R_{{{i+1}}} - {sp.latex(abs(scalar))} R_{{{pivotrow+1}}} "
            elif scalar < 0:
                string += f"R_{{{i+1}}} + {sp.latex(abs(scalar))} R_{{{pivotrow+1}}}"
            
            if i < m - 1:
                string += "\\\ "

            if i == pivotrow:
                continue

            A[i,:] -= scalar * A[pivotrow,:]
            
        if andflag and n < 5:
            string += "\\end{array} & \n"
            andflag = False
        else:
            string += "\\end{array} \\\ \\\ \n"
            andflag = True

        # Print updated matrix
        string += "    \\longrightarrow \n"
        string = print_augmented_matrix(A, string)

        pivotrow += 1

    string += "\\end{align*} $$ \n"
                
    return A, string

# Generate solution string
string = "```{solution} systems-ex-gjelim"  

for sys in systems:
    string += f"\n\n{sys.label} \n"
    A = sys.A
    Ab = A.row_join(sys.b)
    Ab, string = gjelim(Ab, string)

    string += f"\nSo the solution is "
    for i in range(len(sys.b)):
        string += f"$x_{{{i+1}}} = {sp.latex(Ab[i,-1])}$"
        if i < len(sys.b) - 2:
            string += ", "
        elif i == len(sys.b) - 2:
            string += " and "
        else:
            string += "."
            

string += "\n```"
print(string)

subprocess.run("pbcopy", text=True, input=string)
````
